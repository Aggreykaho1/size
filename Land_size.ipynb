```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon
```

```python
file_path = r"C:\Users\HP PAVILION\Downloads\land size.xlsx"
sheet_data = pd.read_excel(file_path, sheet_name='area_mapping_0')
```

```python
sheet_data.head
```

```python
def calculate_area(data_subset):
    coords = list(zip(data_subset['gps_long_final'], data_subset['gps_lat_final']))
    if coords[0] != coords[-1]:  # Ensure the polygon is closed
        coords.append(coords[0])
    polygon = Polygon(coords)
    gdf = gpd.GeoDataFrame(index=[0], geometry=[polygon], crs="EPSG:4326")
    gdf_projected = gdf.to_crs("EPSG:32633")  # Transform to UTM for accurate area
    return gdf_projected.geometry.area.iloc[0]
```

```python
plot_1_data = sheet_data[sheet_data['village_name'].str.startswith('P')]
plot_2_data = sheet_data[sheet_data['village_name'].str.startswith('B')]
plot_3_data = sheet_data[sheet_data['village_name'].str.startswith('R')]
```

```python
plot_1_area = calculate_area(plot_1_data)
plot_2_area = calculate_area(plot_2_data)
plot_3_area = calculate_area(plot_3_data)
```

```python
plot_1_area_acres = plot_1_area / 4046.86
plot_2_area_acres = plot_2_area / 4046.86
plot_3_area_acres = plot_3_area / 4046.86
```

```python
# Display results in acres
print(f"Plot 1 Area: {plot_1_area_acres:.2f} acres")
print(f"Plot 2 Area: {plot_2_area_acres:.2f} acres")
print(f"Plot 3 Area: {plot_3_area_acres:.2f} acres")
```
